```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, Activation
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from google.colab import drive
drive.mount('/content/drive/')
import pandas as pd

from keras import regularizers
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt


# 주식 정보 및 감성 분석 데이터 파일
df_aapl = pd.read_csv('/content/drive/MyDrive/paper/stock/df_aapl.csv')
df_msft = pd.read_csv('/content/drive/MyDrive/paper/stock/df_msft.csv')
df_avgo = pd.read_csv('/content/drive/MyDrive/paper/stock/df_avgo.csv')

# 필요한 데이터만 남기고 정리
df_aapl_1 = df_aapl[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'vaderScore']]
df_aapl_2 = df_aapl[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finbertScore']]
df_aapl_3 = df_aapl[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finetunedFinbertScore']]

df_msft_1 = df_msft[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'vaderScore']]
df_msft_2 = df_msft[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finbertScore']]
df_msft_3 = df_msft[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finetunedFinbertScore']]

df_avgo_1 = df_avgo[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'vaderScore']]
df_avgo_2 = df_avgo[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finbertScore']]
df_avgo_3 = df_avgo[['date', 'Open', 'High', 'Low', 'Close', 'Volume', 'finetunedFinbertScore']]

# 결측치 드롭
df_aapl_1 = df_aapl_1.dropna()
df_aapl_2 = df_aapl_2.dropna()
df_aapl_3 = df_aapl_3.dropna()

df_msft_1 = df_msft_1.dropna()
df_msft_2 = df_msft_2.dropna()
df_msft_3 = df_msft_3.dropna()

df_avgo_1 = df_avgo_1.dropna()
df_avgo_2 = df_avgo_2.dropna()
df_avgo_3 = df_avgo_3.dropna()
```

```python
# https://github.com/ZainUlMustafa/Stock-Prediction-using-News-Info-Sentiment/blob/master/sentiment_analysis.ipynb
# Creating a function to preprocess the dataset

def createDataset(dataset, target, lookBack=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - lookBack):
        a = dataset[i:(i + lookBack), :]
        dataX.append(a)
        dataY.append(target[i + lookBack])
    return np.array(dataX), np.array(dataY)
#enddef
```

```python
# 로드한 데이터셋
df = df_aapl_2

# Preproccessing the dataset for LSTM
features = df.drop(['date', 'Close'], axis=1).values
target = df['Close'].values

scalerFeatures = MinMaxScaler(feature_range=(0, 1))
scalerTarget = MinMaxScaler(feature_range=(0, 1))

featuresScaled = scalerFeatures.fit_transform(features)
targetScaled = scalerTarget.fit_transform(target.reshape(-1, 1))

lookBack = 3
X, y = createDataset(featuresScaled, targetScaled, lookBack)
print(X[:2])
print(y[:2])

trainSize = int(len(X) * 0.8)
testSize = len(X) - trainSize
trainX, testX = X[0:trainSize, :], X[trainSize:len(X), :]
trainY, testY = y[0:trainSize], y[trainSize:len(y)]

trainX = np.reshape(trainX, (trainX.shape[0], lookBack, trainX.shape[2]))
testX = np.reshape(testX, (testX.shape[0], lookBack, testX.shape[2]))
```

```python
import numpy as np
# Calculating the score and accuracy of the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 성능 지표 저장을 위한 리스트
trainMAEs = []
testMAEs = []
trainRMSEs = []
testRMSEs = []
trainR2s = []
testR2s = []

batchSize = 32
epoch = 30
neurons = 100
dropout = 0.2

# 30번 반복 실행
for i in range(30):
    # 모델 구성 및 컴파일
    model = Sequential()
    model.add(LSTM(neurons, return_sequences=True, activation='tanh', input_shape=(lookBack, features.shape[1]), name='lstm_1'))
    model.add(Dropout(dropout, name='dropout_1'))
    model.add(LSTM(neurons, return_sequences=True, activation='tanh', name='lstm_2'))
    model.add(Dropout(dropout, name='dropout_2'))
    model.add(LSTM(neurons, activation='tanh', name='lstm_3'))
    model.add(Dropout(dropout, name='dropout_3'))
    model.add(Dense(units=1, activation='linear', activity_regularizer=regularizers.l1(0.00001)))
    model.add(Activation('tanh', name='activation_1'))

    model.compile(loss='mean_squared_error', optimizer='adam')

    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    model.fit(trainX, trainY, batch_size=batchSize, epochs=epoch, verbose=1, validation_split=0.2, callbacks=[early_stopping])

    # 예측
    trainPredict = model.predict(trainX)
    testPredict = model.predict(testX)

    # 예측값 역정규화
    trainPredictInv = scalerTarget.inverse_transform(trainPredict)
    trainYInv = scalerTarget.inverse_transform(np.reshape(trainY, (trainY.shape[0], 1)))
    testPredictInv = scalerTarget.inverse_transform(testPredict)
    testYInv = scalerTarget.inverse_transform(np.reshape(testY, (testY.shape[0], 1)))

    # MAE 계산
    trainMAEs.append(mean_absolute_error(trainYInv[:, 0], trainPredictInv[:, 0]))
    testMAEs.append(mean_absolute_error(testYInv[:, 0], testPredictInv[:, 0]))

    # RMSE 계산
    trainRMSEs.append(np.sqrt(mean_squared_error(trainYInv[:, 0], trainPredictInv[:, 0])))
    testRMSEs.append(np.sqrt(mean_squared_error(testYInv[:, 0], testPredictInv[:, 0])))

    # R² Score 계산
    trainR2s.append(r2_score(trainYInv[:, 0], trainPredictInv[:, 0]))
    testR2s.append(r2_score(testYInv[:, 0], testPredictInv[:, 0]))

# 평균 성능 지표 계산
print(f'Average Training MAE: {np.mean(trainMAEs)}')
print(f'Average Testing MAE: {np.mean(testMAEs)}')
print(f'Average Training RMSE: {np.mean(trainRMSEs)}')
print(f'Average Testing RMSE: {np.mean(testRMSEs)}')
print(f'Average Training R² Score: {np.mean(trainR2s)}')
print(f'Average Testing R² Score: {np.mean(testR2s)}')
```
